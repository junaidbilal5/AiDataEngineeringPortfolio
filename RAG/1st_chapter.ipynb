## **RAG IMPLEMENTATION WITH YOUR OWN TEXT DATA**

# *step 1 for preparing doc for your text*

```python
from langchain_core.documents import Document
```

```python
my_text = """The capital of France is Paris. 
It is known for its rich history, culture, and iconic landmarks such as the Eiffel Tower and the Louvre Museum."""
```

```python
docs = [Document(page_content=my_text, metadata={"source":"Organization_name","documentID" :"Doc1"})]
```

```python
docs
```

```python
docs[0]
```

# *step 2 splitting your text into chunks*

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
spliter=RecursiveCharacterTextSplitter(chunk_size=50,
                                    chunk_overlap=2)
chunks=spliter.split_documents(docs)
```

```python
chunks
```

# *step 3 creating Embeddings for chunks*

```python
from langchain_openai import OpenAIEmbeddings
```

```python
#openaiembeddings=OpenAIEmbeddings(model="text-embedding-3-small")

embading_model=OpenAIEmbeddings(model="text-embedding-3-small")
#by default it will select with 1536 features
```

```python
embading_model
```

```python
embading_model.embed_query("What is the capital of France?")
```

# *step 4 create and storing embeding into vector store*


lan-chain to heavy left
when we create any vectore store then it will create the embeding for our each chunk and store into vectore store automaically
 

```python
from langchain_community.vectorstores import Chroma
```

```python
vector_store=Chroma.from_documents(documents=chunks, embedding=embading_model)
```

```python
print(vector_store)
```

```python
vector_store.similarity_search("What is the capital of France?", k=2)
```

# *step 5 semantic search*


```python
context=vector_store.similarity_search("What is the capital of France?", k=2)
context
```

# * Talk to llm with context

```python
response=llm.invoke(f"what is the capital germany? you can answer me in following context: {context}")
print(response.content)
```
